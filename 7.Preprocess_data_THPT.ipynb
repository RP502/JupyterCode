##### 3.2.Delete columns (Unnamed: 0, ma_ngoai_ngu)

```python
final_data.drop(columns=['Unnamed: 0', 'ma_ngoai_ngu'], inplace=True)
final_data.head(3)
```

##### 3.3.Rename columns to Vietnamese

```python
final_data.rename(columns={
    'sbd': 'SBD',
    'toan': 'Toán',
    'ngu_van': 'Ngữ Văn',
    'ngoai_ngu': 'Ngoại Ngữ',
    'vat_li': 'Vật Lý',
    'hoa_hoc': 'Hoá Học',
    'sinh_hoc': 'Sinh Học',
    'lich_su': 'Lịch Sử',
    'dia_li': 'Địa Lý',
    'gdcd': 'GDCD',   
}, inplace=True)

```

```python
final_data.head(3)
```

##### 3.4.Convert to float

```python
# Turn score from object type to float type
for col in ['Toán', 'Ngữ Văn', 'Ngoại Ngữ', 'Vật Lý', 'Hoá Học', 'Sinh Học', 'Lịch Sử', 'Địa Lý', 'GDCD']:
    final_data[col] = final_data[col].astype(float)

```

##### 3.5.Check duplicated

```python
final_data[final_data.duplicated()]
```

```python
# Drop duplicated
final_data.drop_duplicates(inplace = True)
```

```python
#check duplicated
final_data[final_data.duplicated()]
```

##### 3.6.Clean SBD

```python
# Clean sbd column
def convert(x):
    x = str(x)
    if len(x) == 7:
        return '0' + x
    return x
```

```python
final_data['SBD'] = final_data['SBD'].apply(lambda x: convert(x))
```

```python
final_data.head(3)
```

##### 3.7.Add KHTN and KHXH columns
###### KHTN = (vat_ly + hoa_hoc + sinh_hoc) / 3
###### KHXH = (lich_su + dia_ly + gdcd) / 3

```python
final_data.columns
```

```python
final_data['KHTN'] = round((final_data['Vật Lý'] + final_data['Hoá Học'] + final_data['Sinh Học']) / 3, 2)
final_data['KHXH'] = round((final_data['Lịch Sử'] + final_data['Địa Lý'] + final_data['GDCD']) / 3, 2)
```

```python
final_data.head(3)
```

##### 3.8.Import province

```python
import geopandas as gpd
map_df = gpd.read_file('./diaphantinh.geojson')
map_df.head()
```

```python
# Fix some typo errors
map_df.loc[44, 'ten_tinh'] = 'Quảng Bình'
map_df.loc[31, 'ten_tinh'] = 'Kiên Giang'
map_df.loc[12, 'ten_tinh'] = 'Cần Thơ'
```

##### 3.9.Create a dictionary to map in VietNam

```python

code_provinces = ['51', '52', '18', '11', '60', '19', '56', '37', '44', '43', '47', '61', '55',
                  '06', '04', '40', '63', '62', '48', '50', '38', '05', '24', '01', '30', '21',
                  '03', '64', '23', '22', '41', '54', '36', '07', '42', '10', '08', '49', '25',
                  '29', '27', '45', '15', '39', '31', '34', '35', '17', '32', '59', '14', '46',
                  '26', '12', '28', '33', '53', '02', '58', '09', '57', '16', '13']
                  
mapping = {code_provinces[i]: map_df['ten_tinh'].unique().tolist()[i] for i in range(len(code_provinces))}
```

##### 3.10.Create columns of code and province

```python

final_data['code'] = final_data['SBD'].str[:2]
final_data['province'] = final_data['code'].apply(lambda x: mapping[x])
```
